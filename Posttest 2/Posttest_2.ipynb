```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="/content/flashscore.csv"
df=pd.read_csv(path)
df.head(5)
```

Statistik Deskriptif untuk kolom numerik (Rank, Points, Tournaments)

```python
desc = df[['Rank', 'Points', 'Tournaments']].describe().T
desc['Q1'] = df[['Rank', 'Points', 'Tournaments']].quantile(0.25)
desc['Q2'] = df[['Rank', 'Points', 'Tournaments']].quantile(0.50)  # median
desc['Q3'] = df[['Rank', 'Points', 'Tournaments']].quantile(0.75)
desc[['mean', 'std', 'min', 'max', 'Q1', 'Q2', 'Q3']]
```

Tampilkan 10 record pertama hanya dengan kolom Rank, Points, **Tournaments**

```python
df[['Rank', 'Points', 'Tournaments']].head(10)
```

 Informasi Dataset

```python
jumlah_record = df.shape[0]
jumlah_atribut = df.shape[1]
nama_atribut = df.columns.tolist()
jumlah_numeric = df.select_dtypes(include='number').shape[1]

print("Jumlah record:", jumlah_record)
print("Jumlah atribut:", jumlah_atribut)
print("Nama atribut:", nama_atribut)
print("Jumlah atribut numeric:", jumlah_numeric)
```

# Visualisasi Data

# Histogram distribusi Point

```python
df['Points'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title("Distribusi Poin Pemain")
plt.xlabel("Points")
plt.ylabel("Jumlah Pemain")
plt.show()
```

#Pie chart 5 negara dengan pemain terbanyak

```python
df['Nationality'].value_counts().head(5).plot(
    kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("5 Negara dengan Pemain Terbanyak")
plt.ylabel("")
plt.show()
```

#Scatter plot Tournament vs Points

```python
plt.scatter(df['Tournaments'], df['Points'], alpha=0.6, color='green')
plt.title("Hubungan Jumlah Turnamen vs Poin")
plt.xlabel("Tournaments")
plt.ylabel("Points")
plt.show()
```

#Boxplot distribusi poin

```python
sns.boxplot(x=df['Points'])
plt.title("Sebaran Poin Pemain")
plt.show()

```

#Korelasi Dataset (Heatmap)


```python
corr = df[['Rank', 'Points', 'Tournaments']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korelasi Antar Variabel Numerik")
plt.show()
```

#Kesimpulan Analisis Dataset flashscore.csv

Berdasarkan hasil analisis data tabular yang dilakukan terhadap dataset flashscore.csv, dapat diperoleh beberapa informasi penting mengenai distribusi data, pola hubungan antar variabel, serta gambaran umum dari atribut numerik maupun kategorikal. Analisis ini membantu memahami bagaimana pemain tersebar, perolehan poin, hingga hubungan jumlah turnamen dengan peringkat. Berikut kesimpulan yang dapat ditarik:




1.   Statistik Deskriptif
Nilai rata-rata, standar deviasi, nilai minimum, maksimum, serta kuartil (Q1, Q2, Q3) berhasil dihitung untuk atribut numerik (Rank, Points, Tournaments). Hasil ini memberikan gambaran umum tentang sebaran data dan adanya outlier pada pemain tertentu dengan poin sangat tinggi.

2.   Informasi Dataset
Dataset terdiri dari sejumlah record dan atribut dengan kombinasi data numerik dan kategorikal. Atribut numerik meliputi Rank, Points, dan Tournaments, sementara atribut kategorikal meliputi Name dan Nationality.


3.   Distribusi Pemain
Dari sisi kategorikal, jumlah label unik menunjukkan banyaknya pemain yang berbeda (Name) dan beragamnya asal negara (Nationality). Beberapa negara mendominasi jumlah pemain terbanyak.

4.   Visualisasi Data


*   Histogram menunjukkan distribusi poin pemain yang mayoritas berada pada rentang menengah, dengan sedikit pemain ber-poin sangat tinggi.
*   Pie chart memperlihatkan 5 negara dengan jumlah pemain terbanyak, yang menggambarkan dominasi negara tertentu dalam dataset.
* Scatter plot antara Tournaments dan Points menunjukkan adanya kecenderungan: semakin banyak turnamen yang diikuti, semakin besar peluang pemain memiliki poin tinggi
* Boxplot membantu mengidentifikasi adanya outlier dalam perolehan poin.


5. Korelasi Data
Heatmap korelasi memperlihatkan hubungan antar variabel numerik. Terlihat bahwa Rank memiliki korelasi yang cukup kuat dengan Points, yang sesuai dengan logika peringkat: semakin tinggi poin, maka semakin baik peringkat pemain.
